```python
import re

trials = 0

pwd = 2021

pat = "\d{4}"

while trials <= 2:
   	try:
   		t = int(input("Enter password: "))
   		if t == pwd:
   			print("access granted")
   			break
   		elif re.search(pat, str(t)) == None:
   			print("only 4 digits allowed!")
   			trials += 1
   			print("Trial N°", trials)
   		else:
   			print("Access denied")
   			trials += 1
   			print("Trial N°", trials)
   	
   	except ValueError:
   		print("Invalid input! only digits allowed!")
   		trials += 1
   		print("Trial N°", trials)
```
